```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
np.random.seed(42)
x = np.random.rand(1000, 1)
k = np.random.rand(1000, 1)
y = 3 * x + k
```

```python
data = pd.DataFrame(data=np.hstack((x, y)), columns=["x", "y"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data[['x']], data['y'], test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual values', alpha=0.5)
plt.scatter(X_test, y_pred, color='red', label='Predicted values', alpha=0.5)
plt.plot(X_test, y_pred, color='orange', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Predictions')
plt.legend()
plt.show()
```
