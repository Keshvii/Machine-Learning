```python
import tensorflow as tf
from tensorflow import keras #API:keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train, y_train),(X_test, y_test) = keras.datasets.mnist.load_data() #downloading dataset
```

```python
X_train, X_test = X_train/255.0, X_test/255.0
#Neural netork work efficiently after Normalization  #Datapreprocessing #[0,255] -> [0,1]
```

```python
model = Sequential() #building block
model.add(Flatten(input_shape = (28,28))) #1-D
model.add(Dense(128, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))#Activation: triggers neurons in our model
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = 'Adam', metrics=['accuracy'])
history = model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```
